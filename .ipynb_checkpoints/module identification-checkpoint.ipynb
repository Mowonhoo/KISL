```python
# -*- coding: UTF-8 -*-
import os, re
import pandas as pd
import numpy as np

from functools import reduce
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.manifold import TSNE
import umap.umap_ as umap

import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
plt.rcParams['font.sans-serif'] = ['Times New Roman']
plt.rcParams['axes.unicode_minus'] = False
matplotlib_axes_logger.setLevel('ERROR')
import time
import warnings
warnings.filterwarnings("ignore")

from labels2color import Labels2Color
from topological_overlap_measure import TOMsimilarity
from visualization import *
from utils import *

mlDir = "./outdir/factor analyzer/factor analyzer"
data_indir = "./outdir/feature selection"
out_dir = "./outdir/module identification"
dcorrDir = './outdir/calculate dcorr coef'
filename = 'BLCA.tcga_gtex.tpm.updown.feature_selection.csv'
K_sequence = range(5, 7, 1)
power = 20
distance = "dcorr"  # "pearson" #

prefix = filename.split('.')[0]
outdir = '/'.join([out_dir, distance])

if not os.path.exists(os.path.join(outdir, prefix)):
    os.makedirs(os.path.join(outdir, prefix))

# Construct GCN
datExpr = pd.read_csv(os.path.join(data_indir, prefix, filename), sep=',', index_col=0)
if distance == "pearson":
    print("Input: datExpr.shape", datExpr.shape)
    print("NAN check for datExpr: ", datExpr.isnull().values.sum())
    sim = datExpr.T.corr()
elif distance == "snn":
    print("Input: datExpr.shape", datExpr.shape)
    print("NAN check for datExpr: ", datExpr.isnull().values.sum())
    sim = snn_sim_matrix(datExpr, k=5)
elif distance == "dcorr":
    sim = pd.read_csv(os.path.join(dcorrDir, prefix, filename.rstrip('.csv') + ".dcorr.csv"), sep=',')
    sim.index = sim.columns
else:
    print("Input: datExpr.shape", datExpr.shape)
    print("NAN check for datExpr: ", datExpr.isnull().values.sum())
    sim = datExpr.T.corr()
    print("input parameter error：Undefined distance type.Adjust to use the default pearson parameter")

print("NAN check: ", sim.isnull().values.sum())
if sim.isnull().values.ravel().sum():
    sim.drop(sim.columns[sim.isnull().all(axis=1)].tolist(), inplace=True)
    sim = sim[sim.columns[~sim.isnull().all(axis=0)].tolist()]
    sim.fillna(0, inplace=True)
print("After processing NA: sim.shape", sim.shape)
sim.to_csv(os.path.join(outdir, prefix, filename.rstrip('csv') + distance + ".sim.csv"), index=False, sep=',')

sft = pickSoftThreshold(sim, power=power, RsquaredCut=0.85, prefix=prefix, outdir=os.path.join(outdir, prefix))
print('powerEstimate:%s' % sft["powerEstimate"])
sft["fitIndices"].to_csv(os.path.join(outdir, prefix, filename.rstrip('csv') + "sft_fitIndices.csv")
                         , index=False, sep=',')

# Check scale free topology
# Create an adjacency network； ^power 
# here we define the adjacency matrix using soft thresholding with beta=6
ADJ = abs(sim) ** sft["powerEstimate"]
print("Input: ADJ", "Max: {}".format(ADJ.values.ravel().max()), "Min: {}".format(ADJ.values.ravel().min()),
      sep="\n")
print("NAN check: ", ADJ.isnull().values.sum())
ADJ.to_csv(os.path.join(outdir, prefix, filename.rstrip('csv') + distance + ".ADJ.csv"), index=False, sep=',')

# Compute topological overlap
TOM = TOMsimilarity(ADJ)
dissTOM = 1 - TOM

TOM.to_csv(os.path.join(outdir, prefix, filename.rstrip('csv') + distance + ".TOM.csv"), index=False, sep=',')
dissTOM.to_csv(os.path.join(outdir, prefix, filename.rstrip('csv') + distance + ".dissTOM.csv"), index=False,
               sep=',')

X = dissTOM.copy()

try:
    X_Dim = umap.UMAP(n_components=3).fit_transform(datExpr)

except Exception as e:
    print("\n==============================")
    print("datExpr dimension reduction failed.\n")
    print(e)
    print("==============================\n")
    X_Dim = TSNE(n_components=3, learning_rate=100, random_state=0).fit_transform(datExpr)
```

```python
# Calculate the learning curve of parameter K
print("Calculate the learning curve for K ==========>")

score = {'SI': [], 'CH': [], 'DBI': []}

df = open(os.path.join(mlDir, prefix, prefix + ".ml.txt"), 'r')
geneslist = df.readline().split("\t")
geneslist.pop()
ml = set()
byt = df.readlines()
for index, value in enumerate(byt):
    temp = value.split("\t")
    temp.pop()
    byt[index] = temp
    for i in range(len(temp)):
        for j in range(i + 1, len(temp)):
            ml.add((geneslist.index(temp[i]), geneslist.index(temp[j])))
print("Number of pairwise constraints:%d\n" % len(ml))

print("K = ", list(K_sequence))
for k in K_sequence:
    print("---->K = %d" % k)
    try:
        t1 = time.time()
        pckm = PCKMeans(n_clusters=k)  # , distance_type='euclidean'
        pckm.fit(np.array(X), ml=ml)
        t2 = time.time()
        print ("the time of clustering is %.5fs" % (t2 - t1))
        pckm_labels = pckm.labels_

        SI = silhouette_score(X, pckm_labels)
        CH = calinski_harabasz_score(X, pckm_labels)
        DBI = davies_bouldin_score(X, pckm_labels)
        score["SI"].append(SI)
        score["CH"].append(CH)
        score["DBI"].append(DBI)
        print("The average silhouette_score is :%.4f" % SI)

    except Exception as e:
        print("\n==============================")
        print("    Error processing %s" % k)
        print("    ", e)
        score["SI"].append(0)
        score["CH"].append(0)
        score["DBI"].append(0)

        print("==============================\n")
        continue

default_K = K_sequence[score["SI"].index(np.max(score["SI"]))]
print('The optimal K value:%d' % default_K)


pd.DataFrame.from_dict({'K_sequence': list(K_sequence)
                           , 'SI': score["SI"]
                           , "CH": score["CH"]
                           , "DBI": score["DBI"]
                           , 'type': ['KISL'] * len(K_sequence)}
                       ).to_csv(os.path.join(outdir, prefix, "%s.K_learn-curve.csv" % filename.rstrip('.csv')),
                                index=False, sep=',')

# Visualization of the learning curve
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(6, 6)

xticks = list(K_sequence)
ax.plot(xticks, score["SI"]
         , color='green', linewidth=1.0, linestyle='--', label='KISL', marker='^', markerfacecolor='green',
         markersize=10)
ax.set_xlabel('K', fontdict={'size': 16})
ax.set_ylabel('Silhouette Coefficient Score', fontdict={'size': 16})
ax.set_title("SI", fontdict={'size': 18})
ax.legend(loc='best')

plt.title(("Learning curve for K values of KISL clustering on %s data" % prefix), fontsize=20,
             fontweight='bold')
fig = plt.gcf()
fig.savefig(os.path.join(outdir, prefix, "%s.Learning curve for K values of KISL clustering on %s data.pdf" % (
    filename.rstrip('.csv'), filename.split('.')[0])), bbox_inches='tight')  #
fig.savefig(os.path.join(outdir, prefix, "%s.Learning curve for K values of KISL clustering on %s data.png" % (
    filename.rstrip('.csv'), filename.split('.')[0])), dpi=1080, bbox_inches='tight')  #
plt.show()
plt.close()
```

```python
print("KISL==========>")

K = default_K if default_K > 3 else K_sequence[score["SI"].index(np.max(score["SI"][1:]))]

t1 = time.time()
pckm = PCKMeans(n_clusters=K)  # , distance_type='euclidean'
pckm.fit(np.array(X), ml=ml)  #
t2 = time.time()
print("the time of clustering is %.5fs" % (t2 - t1))
pckm_labels = pckm.labels_

print("value_counts:\n", pd.DataFrame(pd.value_counts(pckm_labels)).T)
labels_counts = pd.DataFrame(pd.value_counts(pckm_labels, sort=False, ascending=False, normalize=False))
labels_counts.insert(0, "clusters", labels_counts.index)
labels_counts.columns = ["clusters", "counts"]
labels_counts.to_csv(
    os.path.join(outdir, prefix, filename.rstrip('csv') + distance + ".KISL_clustering.labels_counts.csv"),
    index=False)
pd.DataFrame(np.array([X.index, pckm_labels]).T, columns=["gene", "labels"]).to_csv(
    os.path.join(outdir, prefix, filename.rstrip('csv') + distance + ".KISL_clustering.cluster_labels.csv"),
    index=False)

# The visualization of the clustered data ###

plt.clf()
fig = plt.figure(figsize=(6, 6), dpi=100)
fig.set_size_inches(18, 7)
plt.rcParams.update({'font.family': 'Times New Roman'})
plt.rcParams.update({'font.weight': 'normal'})
plt.rcParams.update({'font.size': 20})

colors = Labels2Color(X, pckm_labels)['colorcode']

ax = fig.add_subplot(1, 1, 1, projection='3d')
# ax.axis('off')
ax.scatter(X_Dim[:, 0], X_Dim[:, 1], X_Dim[:, 2]  # 3D
           , marker='o'
           , s=4
           , c=colors
           )
ax.set_title("KISL with n_clusters = %d" % K, fontsize=18)
ax.set_xlabel("The 1st feature", fontsize=16)
ax.set_ylabel("The 2nd feature", fontsize=16)
ax.set_zlabel("The 3rd feature", fontsize=16)

# ax.spines['top'].set_visible(False)
# ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
# ax.spines['left'].set_visible(False)
plt.suptitle(("The visualization of %s data clustering results" % prefix), fontsize=20, fontweight='bold')

fig = plt.gcf()
fig.savefig(os.path.join(outdir, prefix
                         , filename.rstrip('csv') + distance +
                         ".the visualization of %s data with KISL results.png" % prefix)
            , bbox_inches='tight', dpi=1080)
fig.savefig(os.path.join(outdir, prefix
                         , filename.rstrip('csv') + distance +
                         ".the visualization of %s data with KISL results.pdf" % prefix)
            , bbox_inches='tight')
plt.show()
# plt.close()
```
